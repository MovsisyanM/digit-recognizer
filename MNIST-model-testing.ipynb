```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import sklearn, sklearn.linear_model, sklearn.multiclass, sklearn.naive_bayes
```

```python
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()
```

## Classical machine learning models

```python
train_images = train_images.reshape(train_images.shape[0], train_images.shape[1] * train_images.shape[2]) / 255.0
test_images = test_images.reshape(test_images.shape[0], test_images.shape[1] * test_images.shape[2]) / 255.0
```

### One vs rest classifier

```python
log_reg = sklearn.linear_model.SGDClassifier()

one_vs_rest_model = sklearn.multiclass.OneVsRestClassifier(log_reg)

one_vs_rest_model.fit(train_images, train_labels)

predictions = one_vs_rest_model.predict(test_images)
```

```python
from sklearn.metrics import classification_report

class_report = classification_report(test_labels, predictions, output_dict=True)
class_report
```

```python
disp = sklearn.metrics.plot_confusion_matrix(one_vs_rest_model, test_images, test_labels)
```

### Bernoulli Naive Bayes

```python
bernoulli_NB = sklearn.naive_bayes.BernoulliNB()

bernoulli_NB.fit(train_images, train_labels)

predictions = bernoulli_NB.predict(test_images)

disp = sklearn.metrics.plot_confusion_matrix(bernoulli_NB, test_images, test_labels)

class_report = classification_report(test_labels, predictions, output_dict=True)
class_report
```

### Multinomial Naive Bayes

```python
multinomial_NB = sklearn.naive_bayes.MultinomialNB()

multinomial_NB.fit(train_images, train_labels)

predictions = multinomial_NB.predict(test_images)

disp = sklearn.metrics.plot_confusion_matrix(multinomial_NB, test_images, test_labels)

class_report = classification_report(test_labels, predictions, output_dict=True)
class_report
```

### Gaussian Naive Bayes

```python
gaussian_NB = sklearn.naive_bayes.GaussianNB()

gaussian_NB.fit(train_images, train_labels)

predictions = gaussian_NB.predict(test_images)

disp = sklearn.metrics.plot_confusion_matrix(gaussian_NB, test_images, test_labels)

class_report = classification_report(test_labels, predictions, output_dict=True)
class_report
```

### Complement Naive Bayes

```python
complement_NB = sklearn.naive_bayes.ComplementNB()

complement_NB.fit(train_images, train_labels)

predictions = complement_NB.predict(test_images)

disp = sklearn.metrics.plot_confusion_matrix(complement_NB, test_images, test_labels)

class_report = classification_report(test_labels, predictions, output_dict=True)
class_report
```

## Convolutional Neural Networks

```python
train_images = train_images.reshape(60000, 28, 28, 1)
test_images = test_images.reshape(10000, 28, 28, 1)
```

```python
train_labels = tf.keras.utils.to_categorical(train_labels)
test_labels = tf.keras.utils.to_categorical(test_labels)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras import layers
```

```python
def create_model():
    model = Sequential([
        layers.Conv2D(128, 3, activation='relu', input_shape=(28, 28, 1)),
        layers.MaxPool2D(2),
        layers.Conv2D(256, 3, activation='relu'),
        layers.MaxPool2D(2),
        layers.GlobalAveragePooling2D(),
        layers.Dense(10, activation='softmax')
    ])
    
    model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
    
    return model
```

### Base CNN

```python
model = create_model()

model.summary()
```

```python
history_base = model.fit(train_images, train_labels, 
          batch_size=128, epochs=10, 
          validation_data=(test_images, test_labels))
```

```python
plt.figure(figsize=(15, 10))
plt.plot(range(10),  history_base.history['loss'], label='Training loss')
plt.plot(range(10),  history_base.history['val_loss'], label='Validation loss')
plt.title('Base CNN training & validation loss')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.plot(range(10),  history_base.history['accuracy'], label='Training accuracy')
plt.plot(range(10),  history_base.history['val_accuracy'], label='Validation accuracy')
plt.title('Base CNN training & validation accuracy')
plt.legend()
plt.show()
```

```python
predictions
```

```python
from sklearn.metrics import confusion_matrix

predictions = np.argmax(model.predict(test_images), axis=-1)

conf_matrix = confusion_matrix(np.argmax(test_labels, axis=-1), predictions)
print(conf_matrix)
```

```python
class_report = classification_report(np.argmax(test_labels, axis=-1), predictions, output_dict=True)
class_report
```

### CNN With LR Scheduling

```python
def scheduler(epoch):    
    initial_lr = 0.001
    lr = initial_lr * np.exp(-0.1 * epoch)
    return lr
```

```python
from tensorflow.keras.callbacks import LearningRateScheduler
lr_scheduler = LearningRateScheduler(scheduler, verbose=1)
```

```python
model = create_model()

history_scheduling = model.fit(train_images, train_labels, 
                  batch_size=128, epochs=10, 
                  validation_data=(test_images, test_labels), callbacks=[lr_scheduler])
```

```python
plt.figure(figsize=(15, 10))
plt.plot(range(10),  history_base.history['loss'], label='Training loss')
plt.plot(range(10),  history_base.history['val_loss'], label='Validation loss')
plt.title('Base CNN training & validation loss')
plt.legend()
plt.show()
```
